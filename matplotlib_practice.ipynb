```python
# import dependencies
import matplotlib.pyplot as plt
```

```python
# Set x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total far in US dollars accumlated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create the plot (MATLAB)
plt.plot(x_axis, y_axis)
```

```python
# Create same line chart with the ax.plt() (object-oriented interface)

fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Creating plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create plot with plt.show()
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Labeling:
# Create the plot and add label for the legend
plt.plot(x_axis, y_axis, label='Boston')

#Create lables for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare ($)")

#Set the y limit between 0 and 45
plt.ylim(0, 45)

#create a title
plt.title("PyBer Fare by Month")

#Add the legend
plt.legend()
```

```python
# Modifying the code to update visually
plt.figure(figsize=(20, 10))
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.xlim(0, 5)
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add a grid.
plt.grid()

# Add the legend.
plt.legend()
```

```python
# Update chart using the object-oriented approach
# Create the plot
ax = plt.axes()
ax.plot(x_axis, y_axis, marker="", color="green", linewidth=2, label='Boston')

# create x and y labels
ax.set_xlabel('Date')
ax.set_ylabel('Fare ($)')

#Create the labels for the x and y axis
ax.set_ylim(0, 45)

#Create a title
ax.set_title("PyBer Fare by Month")

# Add grid
ax.grid()

#Add legend
ax.legend()
```

```python
# Create the plot with a bar chart
plt.bar(x_axis, y_axis, color="green", label='Boston')

# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare ($)")

#Create a title
plt.title("PyBer Fare by Month")

#Add the legend
plt.legend()
```

```python
# Create horizontal bar chart
plt.barh(x_axis, y_axis, color="magenta", label='Boston')
plt.gca().invert_yaxis()

# Create labels for the x and y axes
plt.xlabel("Fare ($)")
plt.ylabel("Date")

#Create a title
plt.title("PyBer Fare by Month")

#Add the legend
plt.legend()
```

```python
# Bar Chart with the object-oriented method
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color="black", label='New York')
```

```python
# Horizontal Bar chart with object-oriented method
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")

# create x and y labels
ax.set_xlabel('Fare ($)')
ax.set_ylabel('Date')

#Invert y axis
ax.invert_yaxis()

#Create a title
ax.set_title("PyBer Fare by Month")

#Add legend
ax.legend()
```

```python
# Scatter Plot practice
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Adding 'o' to turn it from a line chart to a scatter plot
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(y_axis, x_axis, color="red", label="Chicago")
plt.gca().invert_yaxis()

#Add a title and axes lables
plt.title("PyBar Fare by Month")
plt.xlabel("Fare ($)")
plt.ylabel("Date")

#Add a legend
plt.legend()
```

```python
#Adjusting the size by multiplying each data point in the y axis by 3 by iterating through a loop
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)

# Creating bubble charts
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Creating bubble charts, formatting for list comprehension instead of loop
plt.scatter(x_axis, y_axis, s=[i * 3 for i in y_axis])
```

```python
# Scatter plot with the object-oriented interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Bubble chart with the object-oriented interface
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i * 5 for i in y_axis], c="blue", alpha=0.2, linewidth=2, label='Boston')

#Invert y axis
ax.invert_yaxis()

# create x and y labels
ax.set_xlabel('Fare ($)')
ax.set_ylabel('Date')

# create x and y labels
ax.set_xlim(0, 50)

#Create a title
ax.set_title("PyBer Fare by Month")

#Add legend
ax.legend()
```

```python
# Pie Charts with MATLAB approach
plt.pie(y_axis, labels=x_axis)

plt.show()
```

```python
# Formatting pie chart above

#Assign 12 colors for each month
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))

plt.pie(y_axis, 
        explode=explode_values,
        colors=colors,
        labels=x_axis,
        autopct='%.1f%%')

plt.show()
```

```python
# pie chart with object-oriented interface
fig, ax = plt.subplots(figsize=(8, 8))
explode_values = [0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0]
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

ax.pie(y_axis, 
       explode=explode_values,
       colors=colors,
       labels=x_axis,
      autopct='%.1f%%',
       counterclock=False,
      startangle=90)
plt.show()
```
